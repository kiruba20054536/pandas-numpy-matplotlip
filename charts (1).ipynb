```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Histogram

```python
df_1 = pd.DataFrame(np.random.randn(1000,2),columns=['A','D'])
df_2 = pd.DataFrame(np.random.rand(1000,2),columns=['a','b'])
df_1['A'].plot.hist(bins=40, edgecolor = 'k').autoscale(enable=True, axis ='both', tight=True)
plt.show()
```

Bar plot

```python
df_3 = pd.DataFrame(np.random.rand(20,2),columns=['A','D'])
df_4 = pd.DataFrame(np.random.rand(200,3),columns=['B','C','E'])
print(df_3)
print(df_4)
df_3.plot.bar()#BAR 
df_3.plot.bar(stacked=True)#STACKED
df_3.plot.line(y=['A','D'],figsize=(12,4),lw=2)#LINE GRAPH
df_3.plot.area();#LINE GRAPH
df_4.plot.scatter(x='B',y='C')#SCATTER PLOT
plt.show()
```

SCATTER PLOT WITH COOLWARM

```python
df_4.plot.scatter(x='B',y='C', c='E',cmap= 'coolwarm')#SCATTER PLOT
```

```python
df_4.boxplot()#BOXPLOT
```

```python
df_4.plot.kde();#Kernel Density Estimate (KDE) plot
df_4.plot.density()
```
