CORRELATION COEFFICIENT BETWEEN THE STOCKS OF SNAPCHAT AND META

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = yf.download(["META", "SNAP"], start="2024-01-01", end="2025-01-01")[["Close"]]
data.columns = data.columns.droplevel(0)
data.columns = ["Meta", "Snapchat"]
```

```python
data = data.dropna()

returns = data.pct_change().dropna()


correlation = returns.corr()
print("Correlation Matrix:\n", correlation)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(returns["Meta"], returns["Snapchat"], alpha=0.5, label="Daily Returns")



plt.title("Meta vs Snapchat Daily Returns with Trend Line")
plt.xlabel("Meta Daily Return")
plt.ylabel("Snapchat Daily Return")
plt.grid(True)
plt.legend()
plt.show()
```
