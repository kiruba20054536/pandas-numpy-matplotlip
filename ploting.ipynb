PLOTTING IN PYTHON

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\mccmrfcs\Downloads\AirPassengers.csv')
df
```

```python
plt.plot(df.Month,df['Passengers'])
plt.title("month vs passengers")
plt.xlabel("month")
plt.ylabel("Passengers")
plt.show()
```

```python
plt.figure(figsize=(10, 4), dpi=300)#
plt.plot(df.Month[:50], df['Passengers'][:50])#slicing first 50 month and 50 passengers
plt.xticks(rotation='vertical');
plt.xlabel("year")
plt.ylabel("passenger count")
plt.title("Cropped passenger data")
```

```python
matrix = np.random.rand(10,10)
plt.figure(figsize=(10,8))
plt.imshow(matrix, cmap='viridis', interpolation = 'nearest')
plt.colorbar(label='Value')
plt.title("Matrix heatmap")
plt.show()
```
