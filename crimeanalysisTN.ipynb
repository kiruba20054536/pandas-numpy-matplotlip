```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\mccmrfcs\Downloads\tn-districts-total-complaints-2021-1.csv")
df
```

```python
# Remove 's.no' column
df.drop(columns=['s.no'], inplace=True, errors='ignore')
# Remove the 'Total' row before sorting
df_1 = df.drop(index=46).reset_index(drop=True)
df_1
```

# 1. Top 10 Districts by Total Complaints

```python
top_10 = df_1.sort_values(by="Total Complaints", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x="Total Complaints", y="District", data=top_10, palette="Reds_r")
plt.title("Top 10 Districts by Total Complaints", fontsize=16)
plt.xlabel("Total Complaints")
plt.ylabel("District")
plt.tight_layout()
plt.show()
```

# 2. Comparison of Oral, Written, and Electronic Complaints by District

```python
plt.figure(figsize=(15, 7))
df_compare = df_1[['District', 'Total oral complaints received', 'Total written complaints received', 'Electronic Form']]
df_compare.set_index('District').head(10).plot(kind='bar', stacked=False, figsize=(15, 7), colormap='viridis')
top_10 = df_1.sort_values(by="Total Complaints", ascending=False).head(10)
plt.title("Comparison of Oral, Written, and Electronic Complaints by District", fontsize=16)
plt.ylabel("Number of Complaints")
plt.xlabel("District")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# INTERPETATION: Electronic form of complaints is comparatively very low even in modern cities like  chennai.

# 3. Correlation Between Complaint Types

```python
plt.figure(figsize=(12, 10))
corr = df_1.drop(columns=['District']).corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Complaint Types", fontsize=16)
plt.tight_layout()
plt.show()
```
